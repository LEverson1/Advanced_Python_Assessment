```python
# Web Scraping
from bs4 import BeautifulSoup
import requests
from datetime import datetime
def getHTML(url):
    response = requests.get(url)
    return response.text

base_url = "https://www.timeanddate.com/holidays/us/"

year = "2021"

response = requests.get(base_url + year)

scraped_holidays = []

html = getHTML("https://www.timeanddate.com/holidays/us/")
soup = BeautifulSoup(html,'html.parser')
table = soup.find('table',attrs = {'class':'table table--left table--inner-borders-rows table--full-width table--sticky table--holidaycountry'})
try:
    for row in table.find_all_next('tr'):
        tds = row.find_all_next('td')
        names = row.find_all_next('a')
        dates = row.find_all_next('th',attrs = {'class':'nw'})
        new_date = dates[0].text + ", " + year
        holiday = {}
        holiday['name'] = names[0].text
        holiday['date'] = new_date
        scraped_holidays.append(holiday)
except:
    print("End of list.")
    

for i in scraped_holidays:
    dt_practice = datetime.strptime(i['date'], '%b %d, %Y').date()
    dt_practice1 = datetime.strftime(dt_practice, '%Y-%m-%d')
#     print(dt_practice1)
    i['date'] = dt_practice1
```
